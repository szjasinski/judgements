```python

def prawomocność_func(x):
    if x == "nieprawomocne":
        return 0
    elif x == "prawomocne":
        return 1
    else:
        return "None"

    
def wynik_func(x):
    # uchylono - pozytywny wynik
    # oddalono - negatywny wynik
    if "uchylono" in x.lower() or "uchyla" in x.lower():
        return 1
    elif "oddalono" in x.lower() or "oddalonno" in x.lower():
        return 0
    else:
        return "None"



def miesiac_func(x):
    return datetime.strptime(x, '%Y-%m-%d').date().month

def dzien_tygodnia_func(x):
    return datetime.strptime(x, '%Y-%m-%d').date().weekday()



def sędzia_func(x):
    if isinstance(x,str):
        return re.split(" /", x)[0]
    else:
        return "None"

    

# renaming
df = df.rename(columns = {'Data orzeczenia':'Data_orzeczenia'})
df = df.rename(columns = {'Data wpływu':'Data_wpływu'})
df = df.rename(columns = {'Symbol z opisem':'Symbol_z_opisem'})
df = df.rename(columns = {'Hasła tematyczne':'Hasła_tematyczne'})
df = df.rename(columns = {'Skarżony organ':'Skarżony_organ'})
df = df.rename(columns = {'Treść wyniku':'Treść_wyniku'})

# Adding two new columns to the existing dataframe.
# bydefault splitting is done on the basis of single space.
df[['Data_orzeczenia','orz',"Prawomocność"]] = df.Data_orzeczenia.str.split(expand=True)

#selecting relevant columns
df = df[['Data_orzeczenia', 'Sąd', 'Sędziowie', 'Treść_wyniku', 'Prawomocność']]


#applying functions
df['Prawomocność'] = df['Prawomocność'].apply(prawomocność_func)
df["Treść_wyniku"] = df["Treść_wyniku"].apply(wynik_func)

df["Miesiąc_orzeczenia"] = df["Data_orzeczenia"].apply(miesiac_func)
df["Dzien_tygodnia_orzeczenia"] = df["Data_orzeczenia"].apply(dzien_tygodnia_func)


# 
df = df.join(df['Sędziowie'].str.split('\n',expand=True).add_prefix('name_'))

df["name_0"] = df["name_0"].apply(sędzia_func)
df["name_1"] = df["name_1"].apply(sędzia_func)
df["name_2"] = df["name_2"].apply(sędzia_func)
df
```

```python
def get_names():
    names = list(dum0.columns.values) + list(dum1.columns.values) + list(dum2.columns.values)
    return list(set(names))
```

```python
for col_name in get_names():
    if col_name in dum0:
        status0 = dum0.loc[:, col_name]
    if col_name in dum1:
        status1 = dum1.loc[:, col_name]
    if col_name in dum2:
        status2 = dum2.loc[:, col_name]
    

    new_value = status0 | status1 | status2
    
    df[col_name] = new_value
```
